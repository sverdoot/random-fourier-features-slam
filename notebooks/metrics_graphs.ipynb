```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_json('results_b.json')
```

```python
data.
```

```python
keys = ['(1.0, 1.0)', '(2.0, 3.0)', '(3.0, 5.0)', '(4.0, 7.0)', '(5.0, 10.0)']
```

```python
data['range'][10]
```

```python
ape_trans = []
ape_rot = []
rpe_trans = []
rpe_rot = []

for i in [5, 10, 20, 30]:
    for key in keys:
        ape_trans.append(data['bearing'][i][key][0]['ape_trans'])
        ape_rot.append(data['bearing'][i][key][0]['ape_rot'])
        rpe_trans.append(data['bearing'][i][key][0]['rpe_trans'])
        rpe_rot.append(data['bearing'][i][key][0]['rpe_rot'])
```

```python
ape_trans
```

```python
fig, ax = plt.subplots(2, 2, figsize = (20, 15))
metrics = [ape_trans, ape_rot, rpe_trans, rpe_rot]
names = ['ape_trans', 'ape_rot', 'rpe_trans', 'rpe_rot'] 
num_l = [5, 10, 20, 30]

ax[0, 0].plot([1, 2, 3, 4, 5], ape_trans[:5], label = 'num_landmarks = ' + str(5))
ax[0, 0].plot([1, 2, 3, 4, 5], ape_trans[5:10], label = 'num_landmarks = ' + str(10))
ax[0, 0].plot([1, 2, 3, 4, 5], ape_trans[10:15], label = 'num_landmarks = ' + str(15))
ax[0, 0].plot([1, 2, 3, 4, 5], ape_trans[15:20], label = 'num_landmarks = ' + str(20))
ax[0, 0].set_title('APE_trans')
ax[0, 0].set_ylabel('APE_trans')
ax[0, 0].set_xlabel('noise_level')
ax[0, 0].grid('black')
ax[0, 0].legend()

ax[0, 1].plot([1, 2, 3, 4, 5], ape_rot[:5], label = 'num_landmarks = ' + str(5))
ax[0, 1].plot([1, 2, 3, 4, 5], ape_rot[5:10], label = 'num_landmarks = ' + str(10))
ax[0, 1].plot([1, 2, 3, 4, 5], ape_rot[10:15], label = 'num_landmarks = ' + str(15))
ax[0, 1].plot([1, 2, 3, 4, 5], ape_rot[15:20], label = 'num_landmarks = ' + str(20))
ax[0, 1].set_title('APE_rot')
ax[0, 1].set_ylabel('APE_rot')
ax[0, 1].set_xlabel('noise_level')
ax[0, 1].grid('black')
ax[0, 1].legend()

ax[1, 0].plot([1, 2, 3, 4, 5], rpe_trans[:5], label = 'num_landmarks = ' + str(5))
ax[1, 0].plot([1, 2, 3, 4, 5], rpe_trans[5:10], label = 'num_landmarks = ' + str(10))
ax[1, 0].plot([1, 2, 3, 4, 5], rpe_trans[10:15], label = 'num_landmarks = ' + str(15))
ax[1, 0].plot([1, 2, 3, 4, 5], rpe_trans[15:20], label = 'num_landmarks = ' + str(20))
ax[1, 0].set_title('RPE_trans')
ax[1, 0].set_ylabel('RPE_trans')
ax[1, 0].set_xlabel('noise_level')
ax[1, 0].grid('black')
ax[1, 0].legend()

ax[1, 1].plot([1, 2, 3, 4, 5], rpe_rot[:5], label = 'num_landmarks = ' + str(5))
ax[1, 1].plot([1, 2, 3, 4, 5], rpe_rot[5:10], label = 'num_landmarks = ' + str(10))
ax[1, 1].plot([1, 2, 3, 4, 5], rpe_rot[10:15], label = 'num_landmarks = ' + str(15))
ax[1, 1].plot([1, 2, 3, 4, 5], rpe_rot[15:20], label = 'num_landmarks = ' + str(20))
ax[1, 1].set_title('RPE_rot')
ax[1, 1].set_ylabel('RPE_rot')
ax[1, 1].set_xlabel('noise_level')
ax[1, 1].grid('black')
ax[1, 1].legend()
```
